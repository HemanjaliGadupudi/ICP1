```python
# Question 1(a)

s = input("Enter a string: ")
char_list = list(s)
# Removing 'h' and 'o' (middle characters) as per requirement
del char_list[3]
del char_list[3]   # 'o' shifts to index 3 after first deletion
# Reverse the list
char_list.reverse()
result = "".join(char_list)
print("Output:", result)
```

```python
# Question 1(b)
# Take two numbers from user
a = float(input("Enter first number: "))
b = float(input("Enter second number: "))
print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
# Handle division safely
if b != 0:
    print("Division:", a / b)
else:
    print("Division: Cannot divide by zero")
```

```python
# Question 2
sentence = input("Enter a sentence: ")
# According to our Requirement..
# Replace 'python' with 'pythons'
update_sentence = sentence.replace("python", "pythons")
#REPLACE--Its going to replace given new string into old string position
print("Output:", update_sentence)
```

```python
# Question 3
score = float(input("Enter your score: "))
if score >=90:
    grade = "A"
elif score >=80 and score<=89:
    grade = "B"
elif score >=70 and score<=79:
    grade = "C"
elif score >=60 and score<=69:
    grade = "D"
elif score >=0 and score<=59:
    grade = "F"
else:
    grade = "Invalid score"
print("Grade:", grade)
```
